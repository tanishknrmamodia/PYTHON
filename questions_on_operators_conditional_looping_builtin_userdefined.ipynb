```python
#Count the Number of Vowels in a String (Ignoring Case)
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for char in s if char in vowels)

string = input("Enter a string: ")
print("Number of vowels:", count_vowels(string))

```

```python
#Solve a Quadratic Equation (axÂ² + bx + c = 0)
import math

def solve_quadratic(a, b, c):
    discriminant = b**2 - 4*a*c
    if discriminant < 0:
        return "No real roots"
    elif discriminant == 0:
        root = -b / (2*a)
        return f"One root: {root}"
    else:
        root1 = (-b + math.sqrt(discriminant)) / (2*a)
        root2 = (-b - math.sqrt(discriminant)) / (2*a)
        return f"Roots: {root1}, {root2}"

a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))
print(solve_quadratic(a, b, c))
```

```python
#Calculate Distance Between Two Points in a 2D Plane
import math

def distance(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)

x1, y1 = map(float, input("Enter x1, y1: ").split())
x2, y2 = map(float, input("Enter x2, y2: ").split())

print("Distance between points:", distance(x1, y1, x2, y2))
```

```python
#Text Analysis (Number of Words, Sentences, etc.)
from collections import Counter
import re

def text_analysis(text):
    words = re.findall(r'\b\w+\b', text)
    sentences = re.split(r'[.!?]', text)
    most_frequent = Counter(words).most_common(1)

    return {
        "Word Count": len(words),
        "Sentence Count": len(sentences) - 1,
        "Average Word Length": sum(len(word) for word in words) / len(words) if words else 0,
        "Most Frequent Word": most_frequent[0][0] if most_frequent else None
    }

text = input("Enter text: ")
result = text_analysis(text)
for key, value in result.items():
    print(f"{key}: {value}")
```

```python
#String Compression (Run-Length Encoding)
def compress_string(s):
    compressed = []
    count = 1

    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1

    compressed.append(s[-1] + str(count))  # Append last character count
    return ''.join(compressed)

string = input("Enter a string: ")
print("Compressed string:", compress_string(string))
```
