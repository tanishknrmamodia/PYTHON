```python
class Shape:
    def __init__(self, color, filled):
        self.color = color
        self.filled = filled

    def get_info(self):
        print(f"Color: {self.color}, Filled: {'Yes' if self.filled else 'No'}")


class Rectangle(Shape):
    def __init__(self, color, filled, length, width):
        super().__init__(color, filled)
        self.length = length
        self.width = width

    def calculate_area(self):
        return self.length * self.width

shape = Shape("Red", True)
shape.get_info()

rectangle = Rectangle("Blue", False, 5, 10)
rectangle.get_info()
print(f"Area of rectangle: {rectangle.calculate_area()}")
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        print(f"Make: {self.make}, Model: {self.model}, Year: {self.year}")


class Car(Vehicle):
    def __init__(self, make, model, year, mileage):
        super().__init__(make, model, year)
        self.mileage = mileage

    def display_info(self):
        super().display_info()
        print(f"Mileage: {self.mileage} kmpl")


vehicle = Vehicle("Toyota", "Camry", 2020)
vehicle.display_info()

car = Car("Honda", "Civic", 2022, 18)
car.display_info()
```
